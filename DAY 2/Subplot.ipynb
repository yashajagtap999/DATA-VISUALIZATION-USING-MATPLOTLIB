```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(5,10,10)
y=x**2
plt.plot(x,y,marker="o",linestyle="-.")
plt.title("Data Visualization")
```

```python
# subplot:-- it allows multiple plot to be plotted inside canvas :--plt.subplot(rows,columns,part)

import matplotlib.pyplot as plt
plt.subplot(2,2,1)
plt.plot(x,x*2)
plt.subplot(2,2,2)
plt.bar(x,y)
plt.subplot(2,2,3)
plt.plot(y,y**2)
plt.subplot(2,2,4)
plt.bar(x*4,y*2)
```
