```python
# matplotlib :---- it is a library used for data visualization developed on 2003

import matplotlib.pyplot as plt
vegetables=["potato","tomato","onion","paneer"]
prices_in_kg=[20,50,100,95]
plt.plot(vegetables,prices_in_kg,marker="^",linestyle=":",linewidth=2.5,color="y")
plt.xlabel("vegetables")
plt.ylabel("prices_in_kg")
plt.grid()
```
